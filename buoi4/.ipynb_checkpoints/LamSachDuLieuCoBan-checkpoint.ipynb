```python
import pandas as pd
```

```python
column_names=['ID', 'name', 'age', 'weight', 'm0006', 'm0612', 'm1218', 'f0006','f0612', 'f1218']
df = pd.read_csv('https://raw.githubusercontent.com/TCSMFrogdance/mrnamanalysis/main/dataworld_giangdayit/patient_heart_rate.csv', names=column_names)
df.head()
```

```python
df.info()
```

```python
df[['fistname', 'lastname']]=df['name'].str.split(expand=True)
df = df.drop('name', axis=1)
df
```

```python
weight = df['weight']

for i in range(0, len(weight)):
    x=str(weight[i])
    if "lbs" in x[-3:]:
        x=x[:-3]
        float_x = float(x)
        y = int(float_x/2.2)
        y = str(y)+"kgs"
        weight[i]=y
df
```

```python
df.dropna(how='all', inplace=True)
df
```

```python
df = df.drop_duplicates(subset=['fistname', 'lastname', 'age', 'weight'])
df
```

```python
df.fistname.replace({r'[^x00-\x7F]+':''}, regex=True, inplace=True)
df.lastname.replace({r'[^x00-\x7F]+':''}, regex=True, inplace=True)
df
```
