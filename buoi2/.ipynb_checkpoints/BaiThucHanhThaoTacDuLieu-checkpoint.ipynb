```python
import pandas as pd
```

```python
df = pd.read_csv("https://query.data.world/s/rxwrus2odpb6cwjddwdjpbb5djcnln")
df.head(5)
```

```python
df = df.drop(['T3', 'L3', 'H3', 'V3', 'S3', 'X3', 'D3', 'N3'], axis=1)
df = df.drop(['T4', 'L4', 'H4', 'V4', 'S4', 'X4', 'D4', 'N4'], axis=1)
df = df.drop(['T5', 'L5', 'H5', 'V5', 'S5', 'X5', 'D5', 'N5'], axis=1)
df = df.drop(['STT'], axis=1)
```

```python
df = df.rename(columns={"T1": "TOAN10", "L1": "LY10", "H1": "HOA10", "S1": "SINH10", "V1": "VAN10", "X1": "SU10", "D1": "DIA10", "N1": "NGOAI NGU10"})
df = df.rename(columns={"T2": "TOAN11", "L2": "LY11", "H2": "HOA11", "S2": "SINH11", "V2": "VAN11", "X2": "SU11", "D2": "DIA11", "N2": "NGOAI NGU11"})
df = df.rename(columns={"T6": "TOAN12", "L6": "LY12", "H6": "HOA12", "S6": "SINH12", "V6": "VAN12", "X6": "SU12", "D6": "DIA12", "N6": "NGOAI NGU12"})
df = df.rename(columns={"GT": "GIOI TINH", "DT": "DAN TOC", "KV": "KHU VUC THI", "KT": "KHOI THI", "DH1": "DIEM THI MON 1", "DH2": "DIEM THI MON 2", "DH3": "DIEM THI MON 3",})
```

```python
df.head(3)
```

```python
df['TBM1'] = (df['TOAN10']*2 + df['LY10'] + df['HOA10'] + df['SINH10'] + df['VAN10']*2 + df['SU10'] + df['DIA10'] + df['NGOAI NGU10'])/10
df['TBM2'] = (df['TOAN11']*2 + df['LY11'] + df['HOA11'] + df['SINH11'] + df['VAN11']*2 + df['SU11'] + df['DIA11'] + df['NGOAI NGU11'])/10
df['TBM3'] = (df['TOAN12']*2 + df['LY12'] + df['HOA12'] + df['SINH12'] + df['VAN12']*2 + df['SU12'] + df['DIA12'] + df['NGOAI NGU12'])/10
df.head(3)
```

```python
df['XL1'] = ''
df['XL2'] = ''
df['XL3'] = ''
```

```python
df.loc[(df['TBM1']) < 5, 'XL1'] = 'Y'
df.loc[(df['TBM1']) >= 5, 'XL1'] = 'TB'
df.loc[(df['TBM1']) >= 6.5, 'XL1'] = 'K'
df.loc[(df['TBM1']) >= 8, 'XL1'] = 'G'
df.loc[(df['TBM1']) >= 9, 'XL1'] = 'XS'
```

```python
df.loc[(df['TBM2']) < 5, 'XL2'] = 'Y'
df.loc[(df['TBM2']) >= 5, 'XL2'] = 'TB'
df.loc[(df['TBM2']) >= 6.5, 'XL2'] = 'K'
df.loc[(df['TBM2']) >= 8, 'XL2'] = 'G'
df.loc[(df['TBM2']) >= 9, 'XL2'] = 'XS'
```

```python
df.loc[(df['TBM3']) < 5, 'XL3'] = 'Y'
df.loc[(df['TBM3']) >= 5, 'XL3'] = 'TB'
df.loc[(df['TBM3']) >= 6.5, 'XL3'] = 'K'
df.loc[(df['TBM3']) >= 8, 'XL3'] = 'G'
df.loc[(df['TBM3']) >= 9, 'XL3'] = 'XS'
```

```python
df.head(15)
```

```python
def normalize(x):
    return x*0.4
```

```python
df['US_TBM1'] = normalize(df['TBM1'])
df['US_TBM2'] = normalize(df['TBM2'])
df['US_TBM3'] = normalize(df['TBM3'])
df.head(5)
```

```python
df['KQXT'] = 'ROT'
```

```python
for khoi in df['KHOI THI']:
  if khoi == 'A1' and khoi == 'A':
    df.loc[((df['DIEM THI MON 1']*2 + df['DIEM THI MON 2'] + df['DIEM THI MON 3'])/4 >= 5), 'KQXT'] = 'DAU'
  elif khoi == 'B':
    df.loc[((df['DIEM THI MON 1'] + df['DIEM THI MON 2']*2 + df['DIEM THI MON 3'])/4 >= 5), 'KQXT'] = 'DAU'
  else:
    df.loc[((df['DIEM THI MON 1'] + df['DIEM THI MON 2'] + df['DIEM THI MON 3'])/3 >= 5), 'KQXT'] = 'DAU'

df.head(5)
```

```python
df.to_csv('ketquaxettuyendaihoc.csv')
```
